```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn

'''
#####################
绘制基本的二维图，核心的关键点在于numpy对数据的处理
#####################
'''
# data = np.arange(10)
# print(type(data), data)
# plt.plot(data)

# fig1 = plt.figure()
# ax1 = fig1.add_subplot(1, 1, 1)
# ax1.plot('x','y', 'g')
# plt.plot([1.5, 3.5, -2, 3])
# plt.plot(np.random.randn(50).cumsum(), 'k--')  # k--是线条样式
# ax1.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))

fig, axes = plt.subplots(2, 2, sharex=False, sharey=False)  # sharex,使用同一x轴

axes[0, 0].scatter(np.arange(50),
                   np.arange(50) + 3 * np.random.randn(50))  # 散点图
axes[0, 0].plot(np.random.randn(50).cumsum(), 'k--')  # 折线图, k--, 样式
axes[0, 0].set(
    # xlim=[0, 50],
    # ylim=[0, 50],
    title='南京各月度数据',
    xlabel='金额',
    ylabel='月份',

)
axes[1, 1].hist(np.random.randn(50), bins=20, color='k', alpha=0.3)  # 柱状图

plt.rcParams['font.sans-serif'] = ['SimHei']  # 显示中文标签
plt.rcParams['axes.unicode_minus'] = False
plt.subplots_adjust(left=None, bottom=None, right=None, top=None,
                    wspace=None, hspace=None)  # 设置自地图布局参数

plt.show()
```
